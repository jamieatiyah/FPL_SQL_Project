# Creating Gameweek Data

```python
def get_gameweek_history(player_id):
    '''get all gameweek info for a given player_id'''
    
    # send GET request to
    # https://fantasy.premierleague.com/api/element-summary/{PID}/
    r = requests.get(
            base_url + 'element-summary/' + str(player_id) + '/'
    ).json()
    
    # extract 'history' data from response into dataframe
    df = pd.json_normalize(r['history'])
    
    return df


# show player #4's gameweek history
get_gameweek_history(4)[
    [
        'round',
        'total_points',
        'minutes',
        'goals_scored',
        'assists'
    ]
].head()
```

```python
from tqdm.auto import tqdm
tqdm.pandas()
```

```python
points = Players_Df['id'].progress_apply(get_gameweek_history)
```

```python
def get_fixtures(player_id):
    '''get all fixture info for a given player_id'''
    
    # send GET request to
    # https://fantasy.premierleague.com/api/element-summary/{PID}/
    r = requests.get(
            base_url + 'element-summary/' + str(player_id) + '/'
    ).json()
    
    # extract 'fixtures' data from response into dataframe
    df = pd.json_normalize(r['fixtures'])
    
    return df
```

```python
get_fixtures(1).head()
```

```python
# get gameweek histories for each player
fixtures = players['id'].progress_apply(get_fixtures)

# combine results into single dataframe
fixtures = pd.concat(df for df in fixtures)
```
